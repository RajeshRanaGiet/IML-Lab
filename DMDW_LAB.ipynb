```python
#now the pandas package can be referred to as pd instead of pandas.
import pandas as pd
mydataset = {
    'cars' : ["BMW","Volvo","Ford"],
    'passings' : [3,7,2]
}

myvar = pd.DataFrame(mydataset)
print(myvar)
```

```python
# a pandas series is like a column in a table.
import pandas as pd
a = [1,7,2]
myvar = pd.Series(a)
print(myvar)
```

```python
#Labels
import pandas as pd
a = [1,7,2]
myvar = pd.Series(a, index=["x","y","z"])
print(myvar)
```
